```python
import pandas as pd
import seaborn as sns
```

```python
confirmados = pd.read_csv('Casos_Diarios_Estado_Nacional_Confirmados_20200527.csv', error_bad_lines=False)
confirmados.set_index('nombre', drop=True, inplace=True)
confirmados = confirmados.T

defunciones = pd.read_csv('Casos_Diarios_Estado_Nacional_Defunciones_20200527.csv')
defunciones.set_index('nombre', drop=True, inplace=True)
defunciones = defunciones.T

sospechosos = pd.read_csv('Casos_Diarios_Estado_Nacional_Sospechosos_20200527.csv')
sospechosos.set_index('nombre', drop=True, inplace=True)
sospechosos = sospechosos.T

negativos = pd.read_csv('Casos_Diarios_Estado_Nacional_Negativos_20200527.csv')
negativos.set_index('nombre', drop=True, inplace=True)
negativos = negativos.T
```

# Confirmados_Cumulativo

```python
confirmados.drop(['cve_ent','poblacion'],axis=0,inplace=True)
confirmados
```

```python
confirmados.cumsum()
```

# Negativos_Cumulativos

```python
negativos.drop(['cve_ent','poblacion'],axis=0,inplace=True)
negativos
```

```python
negativos.cumsum()
```

# Defunciones_Cumulativo

```python
defunciones.drop(['cve_ent','poblacion'],axis=0,inplace=True)
defunciones
```

```python
defunciones.cumsum()
```

```python
sospechosos.drop(['cve_ent','poblacion'],axis=0,inplace=True)
sospechosos
```
